```python
# pulling all links for the profile

cj = http.cookiejar.MozillaCookieJar()
opener = urllib.request.build_opener(urllib.request.HTTPCookieProcessor(cj))

category_list = ['civil-engineers-and-contractors','architects','interior-designers-and-decorators','civil-engineers-and-contractors']
page = "https://www.houzz.in/professionals/"+category_list[0]+"/c/San-Francisco--CA"
print("Page URL : ", page)



sourcecode = opener.open(page).read()
soup = bs.BeautifulSoup(sourcecode)
cj.clear_session_cookies()
# print(soup.prettify())
time.sleep(10)

for titles in soup.findAll("h1", class_="header-2 header-dt-1 main-title"):
    length_of_cat = titles.text.split()[0]
    length_of_cat = length_of_cat.replace(",","")
    
    print("Total PRofiles in Category :",length_of_cat)
n_pages = int(length_of_cat) // 15
print("No. of Pages to crawl :",n_pages)

## For finding list of Contact urls on each category page 
link_list=[]
try:
    
    for nav_pages in range(0, n_pages):
        urls = "https://www.houzz.in/professionals/"+category_list[0]+"/c/San-Francisco--CA/p/"+str(nav_pages*15)
        sourcecode = opener.open(urls).read()
        soup_new = bs.BeautifulSoup(sourcecode)

        for links in soup_new.findAll("div", class_="name-info"):
    #         print(links)
            for link in links.findAll("a"):
                test = link.get("href")
                if test =="javascript:;" or 'browseReviews' in test.split("/") :
                    continue
                else:
                    print("test links" , test)
                    link_list.append(test)
        cj.clear_session_cookies()
#         break
        time.sleep(15)
except Exception as e:
    print("Exception Text :\n",e)
    print("Length of link list :", len(link_list))
    cj.clear_session_cookies()
                

```

```python
# Getting detailed list 

contact_no_list =[]
website_link_list=[]
company_name_list=[]
contact_name_list = []
contact_email_list=[]

combined_list = []

cnt=0
for i in link_list :
    cnt+=1
    print("Contact Count :", cnt)
    print("==========================")
    
    ## Major source code read of profile
    
    try:
        
        sourcecode = opener.open(i).read()
        soup_text = bs.BeautifulSoup(sourcecode)
    except:
        print("Unable to read source code:-",i)
    
    ## Fetching Contact Phone No
    
    try:
        contact_no = soup_text.find("span", class_="pro-contact-text").text
        print("Phone No : ",contact_no)
        
#         contact_no_list.append(contact_no)
        
    except:
        contact_no="Not available"
#         contact_no_list.append(contact_no)
            
    ## Fetching Website URL  
    
    try:    
        website_link=soup_text.find("div",class_="pro-contact-methods one-line trackMe").find("a").get("href")
        print("Website URL : ",website_link)
#         website_link_list.append(website_link)
    except:
        website_link="Not available"
#         website_link_list.append(website_link)
        
    ## Fetching Company Name
    
    try:    
        company_name=soup_text.find("a", class_="profile-full-name").text
        print("Company Name :",company_name)
#         company_name_list.append(company_name)
        
    except:
        company_name="Not available"
#         company_name_list.append(company_name)
        
    ## Fetching Contact Name
    
    Contact_person=""
    try:    
        for i in soup_text.find_all("div",class_='info-list-text'):
#             print("i",i)
            if 'Contact' in i.text:

                print("Contact :",i.text[8:])
                Contact_person = i.text[8:]
#                 contact_name_list.append(Contact_person)
#                 break
    except:
        Contact_person="Not Available"
#         contact_name_list.append(Contact_person)
    
    ## Fetching Email of Contacts
    
    lst=[]
    try:  
        req = Request(website_link, headers={'User-Agent': 'Mozilla/5.0'})
        sourcecode_1 = urlopen(req).read()
#         sourcecode_1 = opener.open(website_link).read()
        soup_text_1 = bs.BeautifulSoup(sourcecode_1)
#         cj.clear_session_cookies()
        var_company= '(\w+@'+re.findall('www\.*\w+', website_link)[0][4:]+'.com)'
#         print(var_company)
        lst = list(set(re.findall( var_company , str(soup_text_1))))
        print("Email :",lst)
        
#         contact_email_list.append(lst)
        
    except Exception as e:
        lst=["Not Available"]
        print("In Email block exception \n", e)
#         contact_email_list.append(lst)
    print("XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX")    
#       print(contact_person)
    time.sleep(10)
    
#     break
    
#     if contact_no == 'Not Available' or Contact_person=="Not Available" or len(lst)==0 or company_name=="Not available" or website_link=="Not available":
#         continue
#     else:
    contact_no_list.append(contact_no) 
    website_link_list.append(website_link)
    company_name_list.append(company_name)
    contact_name_list.append(Contact_person) 
    contact_email_list.append(lst)
        

combined_list.append([contact_no_list,website_link_list,company_name_list,contact_name_list,contact_email_list]) 

header = ['contact_no_list','website_link_list','company_name_list','contact_name_list','contact_email_list']


dataframe = pd.DataFrame(data = combined_list, columns = header)

dataframe.to_html( "Houzz.html" )
                
                
#         link = links.find("a").get("href")
#         link_list.append(link)
#     print(link_list)
print("Length of Link List ",len(link_list))
```
